```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv('data_out.csv')
df2 = pd.read_csv('data_out2.csv')

df = (df1 + df2) / 2
df = df[df1.score < 1.4][df2.score < 1.4]
df20 = df[df.pop_size == 20]
# for t in df.t_size.unique():
df_t = df20.set_index(['t_size', 'm_rate']).unstack()
df_t['score'].plot(kind='bar', figsize=(10, 8), ylabel='Fitness Scores', xlabel='Tournament Size')
plt.legend(title='Max Mutation Numbers')
```

```python
fig, axs = plt.subplots(1,4)
fig.set_size_inches(20,8)
fig.text(0.1,0.5, 'Fitness Scores', rotation='vertical')
for ax, pop in zip(axs, np.sort(df.pop_size.unique())):
    df[df.pop_size == pop].set_index(['t_size', 'm_rate']).unstack()['score'].plot(kind='bar', ax=ax, title=f'Pop size {int(pop)}', ylim=(0,1.2))
```
